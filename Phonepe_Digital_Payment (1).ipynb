## 3.1: Ensure data consistency across state and district levels

### For each state, calculate the total number of transactions, total transaction amount, and 
### total registered users by summing up the values from the district level data.

```python
district_agg = (
    District_Txn_and_users
    .groupby('State', as_index=False)
    .agg(
        District_Transactions=('Transactions', 'sum'),
        District_Amount_INR=('Amount (INR)', 'sum'),
        District_Registered_Users=('Registered Users', 'sum')
    )
)

district_agg
```

```python
state_agg = (
    State_Txn_and_users
    .groupby('State', as_index=False)
    .agg(
        State_Transactions=('Transactions', 'sum'),
        State_Amount_INR=('Amount (INR)', 'sum'),
        State_Registered_Users=('Registered Users', 'sum')
    )
)
state_agg
```

```python
comparison = district_agg.merge(
    state_agg,
    on='State',
    how='inner'
)
comparison
```

```python
comparison['Txn_Diff'] = (
    comparison['District_Transactions'] - comparison['State_Transactions']
)

comparison['Amount_Diff'] = (
    comparison['District_Amount_INR'] - comparison['State_Amount_INR']
)

comparison['Users_Diff'] = (
    comparison['District_Registered_Users'] - comparison['State_Registered_Users']
)
```

```python
#### Display any discrepancies found between the district-level and state-level data.
```

```python
discrepancies = comparison[
    (comparison['Txn_Diff'] != 0) |
    (comparison['Amount_Diff'] != 0) |
    (comparison['Users_Diff'] != 0)
]

discrepancies
```

# Task 4: Data Merging and Advanced Analysis

## 4.1: Ratio of users to population by state

### Merge the State_Txn and Users dataset with the District Demographics dataset to calculate the 
### ratio of registered users to the population for each state. Display the results in a tabular format.

```python
state_population = (
    District_Demographics
    .groupby('State', as_index=False)
    .agg(Total_Population=('Population', 'sum'))
)

state_users = (
    State_Txn_and_users
    .groupby('State', as_index=False)
    .agg(Total_Registered_Users=('Registered Users', 'sum'))
)

state_user_population = state_users.merge(
    state_population,
    on='State',
    how='inner'
)

state_user_population['User_to_Population_Ratio'] = (
    state_user_population['Total_Registered_Users'] /
    state_user_population['Total_Population']
)

state_user_population[['State', 'User_to_Population_Ratio']]
```

### Create a column chart depicting the ratio of users to population by state.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.bar(
    state_user_population['State'],
    state_user_population['User_to_Population_Ratio']
)

plt.xlabel('State')
plt.ylabel('User to Population Ratio')
plt.title('Ratio of Registered Users to Population by State')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## 4.2: Correlate population density with transaction volume

```python
Merge the District_Txn and Users dataset with the District Demographics dataset.
```

```python
district_merged = District_Txn_and_users.merge(
    District_Demographics,
    on=['State', 'District'],
    how='inner'
)
district_merged
```

### Calculate the correlation between population density and transaction volume.

```python
district_merged['Population_Density'] = (
    district_merged['Population'] / district_merged['Area (sq km)']
)

district_summary = (
    district_merged
    .groupby(['State', 'District'], as_index=False)
    .agg(
        Population_Density=('Population_Density', 'mean'),
        Total_Transactions=('Transactions', 'sum')
    )
)

correlation = district_summary[
    ['Population_Density', 'Total_Transactions']
].corr().iloc[0, 1]

print("Correlation between population density and transaction volume:", correlation)
```

### Create a scatter plot to visualize the correlation between population density and transaction volume.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(
    district_summary['Population_Density'],
    district_summary['Total_Transactions'],
    alpha=0.6
)

plt.xlabel('Population Density')
plt.ylabel('Total Transactions')
plt.title('Population Density vs Transaction Volume (District Level)')
plt.tight_layout()
plt.show()
```

### Summarize your findings. [Hint: you can use corr()]

```python
district_merged = District_Txn_and_users.merge(
    District_Demographics,
    on=['State', 'District'],
    how='inner'
)


district_merged['Population_Density'] = (
    district_merged['Population'] / district_merged['Area (sq km)']
)

district_summary = (
    district_merged
    .groupby(['State', 'District'], as_index=False)
    .agg(
        Population_Density=('Population_Density', 'mean'),
        Total_Transactions=('Transactions', 'sum'),
        Total_Amount_INR=('Amount (INR)', 'sum')
    )
)

correlation_volume = district_summary[
    ['Population_Density', 'Total_Transactions']
].corr().iloc[0, 1]

print(correlation_volume)
```

```python
correlation_amount = district_summary[
    ['Population_Density', 'Total_Amount_INR']
].corr().iloc[0, 1]

print(correlation_amount)
```
